```python
x= np.arange(10)
print(x)
```

```python
y = x**2
print(y)
```

```python
plt.plot(x,y)
plt.xlabel("value")
plt.ylabel("Square |of value")
plt.title("Value vs Square of value")
plt.show()
```

```python
y1 = 2*x
y1
```

```python
plt.plot(x,y,label='square', marker='o',linestyle='dashed')
plt.plot(x,y1,label='double',marker="*")
plt.xlabel("value")
plt.ylabel("Square |of value")
plt.title("Value vs Square of value")
plt.legend()
plt.show()
```
