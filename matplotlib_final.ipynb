# Bar Graphs
- plt.bar()
- plot a bar graph between heights and labels you provides.
- Bar charts are good for comparisons

```python
year = np.arange(4)
car1 = [40, 17, 28, 12]
car2 = [34, 21, 24, 15]
```

```python
plt.figure(figsize=(8,6))
plt.bar(year,car1,label='BMW', tick_label= [2021,2022,2023,2024])
plt.bar(year,car2,label='Audi')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in Million Dollar")
plt.title("Cars price comparison")
plt.show()
```

```python
year
```

```python
plt.figure(figsize=(8,6))
plt.bar(year,car1,width=0.3,label='BMW',tick_label= [2021,2022,2023,2024])
plt.bar(year+0.3,car2,width=0.3,label='Audi')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in Million Dollar")
plt.title("Cars price comparison")
plt.show()
```

# Pie Chart
- plt.pie()
- plot the pie chart with the provides labels and values
- pie charts should be used only for simple compositions never for comparisons or distributions.

```python
subjects = ["Physics", 'Chemistry', 'Maths', "English"]
marks = [78, 45, 90, 65]
```

```python
78/sum(marks)
```

```python
plt.figure(figsize=(10,9))
plt.pie(x= marks,labels=subjects,explode=[0.1,0.1,0.1,0.1],shadow=True, autopct="%0.2f",rotatelabels=True)
plt.legend(subjects)
plt.title("subject wise marks distribution")
plt.show()
```

# Histogram
- When you want to explore how data values are spread across different ranges or bins.
- for continuous datapoints
- plt.hist()
- range = (max - min)/bin

```python
freq,bins,_=plt.hist(s['petal_length'],bins=8)
for f,b in zip(freq,bins):
    plt.text(b+0.05,f+0.5,str(int(f)), ha='center', va='top')
plt.show()
```

# Subplots
create multiple plots in just a single figure.

```python
plt.subplot(2,2,1)
plt.scatter(s['sepal_length'],s['sepal_width'],label='setosa')
plt.scatter(v['sepal_length'],v['sepal_width'],label='virginica')
plt.scatter(vr['sepal_length'],vr['sepal_width'],label='versicolor')
plt.legend()
plt.xlabel("sepal width")
plt.ylabel("sepal length")
plt.title("Species wise sepal analysis")

plt.subplot(2,2,2)
plt.pie(df['species'].value_counts(), labels= df['species'].value_counts().index)

plt.subplot(2,2,3)
freq1,bins1,_=plt.hist(df['petal_length'],bins=6, label='petal')
for f,b in zip(freq1,bins1):
    plt.text(b+0.05,f+0.5,str(int(f)), ha='center', va='top')
freq2,bins2,_=plt.hist(df['sepal_length'],bins=6, label='sepal')
for f,b in zip(freq2,bins2):
    plt.text(b+0.05,f+0.5,str(int(f)), ha='center', va='top')
plt.legend()

plt.subplot(2,2,4)
plt.plot('sepal_length','sepal_width', data=df,label='sepal')
plt.plot('petal_length','petal_width', data=df,label='petal')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

a = np.linspace(-5, 5, 100)  # Sample data for the x-axis
b = np.linspace(-5, 5, 100)  # Sample data for the y-axis
a, b = np.meshgrid(a, b)  # Create a meshgrid for a 2D surface

Z = np.array([[6]])  # Example data for the z-axis

ax.plot_surface(a, b, Z)
plt.show()
```
